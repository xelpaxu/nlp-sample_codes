## 2a. (10 points) Extract all of the words starting with an upper case letter from the text:

```python
import re
```

```python
paragraph = """Alice was beginning to get very tired of sitting by her sister on the bank,
and of having nothing to do. Once or twice she had peeped into the book
her sister was reading, but it had no pictures or conversations in it, &quot;and
what is the use of a book,&quot; thought Alice, &quot;without pictures or
conversations?"""
```

```python
pattern = r"[A-Z][a-zA-Z]*"
```

```python
uppercase_words = re.findall(pattern, paragraph)

for i in uppercase_words:
    print(i)
```

## 2b. (10 points) Read the “melville-moby_dick.txt” into a Python program and extract all of the instances of the word whale in said source.

```python
file_path = "melville-moby_dick.txt"
```

```python
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
```

```python
pattern = r"whale"
```

```python
whale_instances = re.findall(pattern, text)

count = len(whale_instances)
print("Whale Instances:", count)
```

## 2c. (10 points) Download the NLTK package into your local environment and import webtext (from nltk.corpus import webtext). Load the pirates.txt and extract all of the lines spoken by the character Jack Sparrow.

```python
from nltk.corpus import webtext
```

```python
pirates_text = webtext.raw("pirates.txt")
```

```python
pattern = r"^JACK SPARROW: .*"
```

```python
jack_lines = re.findall(pattern, pirates_text, flags=re.MULTILINE)

for i in jack_lines:
    print(i)
```
